Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

graph

```python
##draw a line in a diagram from position (0,10) to positon (5,50)
xpoints = np.array([1,2])
ypoints = np.array([10,2])


plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints = np.array([1,3])
ypoints = np.array([2,6])
plt.plot(xpoints,ypoints)
plt.show()
```

graph with marker symbol

```python
xpoints = np.array([1,6])
ypoints = np.array([2,5])
plt.plot(xpoints,ypoints,'x')
plt.show()
```

graph with marker color

```python
ypoints = np.array([2,8,9,3,5])
plt.plot(ypoints,'*''r')
plt.show()
```

with markersize(ms)

```python
xpoints = np.array([1,8,9,7,6])
ypoints = np.array([2,8,9,3,5])
plt.plot(xpoints,ypoints,marker='*',ms=30,color='red')
plt.show()
```

with mec(marker edge color)

```python
ypoints = np.array([2,8,9,3,5])
plt.plot(ypoints ,marker = 'o', ms = 30, mec = 'r', color = 'pink')
plt.show()
```

with mfc(marker face color)

```python
xpoints=np.array([1,2,34,5])
ypoints=np.array([2,3,45,6])

plt.plot(xpoints,ypoints,marker='o',ms=30,mec='k',mfc='green')
plt.show()
```

using both mfc and mec

```python
##USE BOTH MEC AND MFC ARGUMENTS TO COLOR OF ENTIRE MARKER
ypoints=np.array([4,5,6,7,6])
plt.plot(ypoints,marker='o',ms=30,mec='k',mfc='blue')
plt.show()
```

with line style

```python
xpoints=np.array([1,2,34,5])
ypoints=np.array([2,3,45,6])

plt.plot(xpoints,ypoints,marker='o',linestyle='dashed',ms=30,mec='k',mfc='green')
plt.show()
```

with linewidth

```python
xpoints=np.array([1,2,34,5])
ypoints=np.array([2,3,45,6])

plt.plot(xpoints,ypoints,marker='o',linestyle='dashed',linewidth=5,ms=30,mec='k',mfc='green')
plt.show()
```

only with linewidth

```python
### you can use the keyword argument
### the value is a floating number,in points
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linewidth=5)
plt.show()
```

2 graphs in one plot

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,5,11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

label included with fontsize and color

```python
x=np.array([80,85,90,95,100,105,110])
y=np.array([240,250,260,270,280,290,300])
plt.plot(x,y,color="r",linewidth=1)

plt.xlabel("average pulse",fontsize=30,color='red')
plt.ylabel("calorie burnage",fontsize=10,color='g')
plt.show()
```

with title and color

```python
x=np.array([80,85,90,95,100,105,110])
y=np.array([240,250,260,270,280,290,300])
plt.plot(x,y,color="r",linewidth=1)

plt.xlabel("average pulse",fontsize=30,color='red')
plt.ylabel("calorie burnage",fontsize=10,color='g')
plt.title("sample",fontsize=30,color='orange')
plt.show()
```

with grid

```python
x=np.array([80,85,90,95,100,105])
y=np.array([240,250,260,270,280,290])

plt.title("sports watch data")
plt.xlabel("average pulse")
plt.ylabel("calorie burnage")
plt.plot(x,y)
plt.grid(color="green")
plt.show()
```

with gridcolor and line style

```python
##you can use the axis parameter in the grid() function to specify which grid lines to display
##legal values are:'x','y', and 'both'.default value is 'both'
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("sports watch data")
plt.xlabel("average pulse")
plt.ylabel("calorie burnage")
plt.plot(x,y,color="0")
plt.grid(color="green",linestyle="dotted")
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("sports watch data")
plt.xlabel("average pulse")
plt.ylabel("calorie burnage")
plt.plot(x,y)
plt.grid(axis='x')
plt.show()
```

pie chart with labels

```python
x=np.array([1,2,3,4,5])
mylabels=["cherries","apples","grapes","orange","banana"] 
plt.pie(x,labels=mylabels)
plt.show()
```

explode

```python
x=np.array([1,2,3,4,5])
mylabels=["cherries","apples","grapes","orange","banana"] 
myexplode=[0,0.5,0.6,0.7,0.7]
plt.pie(x,labels=mylabels,explode=myexplode)
plt.show()
```

startangle for dividing to exact angle

```python
x=np.array([1,2,3,4,5])
mylabels=["cherries","apples","grapes","orange","banana"] 
myexplode=[0,0.5,0.6,0.7,0.7]
plt.pie(x,labels=mylabels,explode=myexplode,startangle=90)
plt.show()
```

for angle (startangle)

```python
x=np.array([1,2,3,4,5])
mylabels=["cherries","apples","grapes","orange","banana"] 

plt.pie(x,labels=mylabels,startangle=90)
plt.show()
```

colors for pie chart

```python
x=np.array([1,2,3,4,5])
mylabels=["cherries","apples","grapes","orange","banana"] 
mycolors=["hotpink","k","blue","orange","yellow"]

plt.pie(x,labels=mylabels,colors=mycolors,startangle=90)
plt.show()
```
