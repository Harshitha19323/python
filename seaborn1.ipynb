Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

reading csv file

```python
df = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/HHE_14th-Aug//main/loan_prediction.csv")
```

data types

```python
df.dtypes
```

head

```python
df.head()
```

relplot

```python
sns.relplot(x="ApplicantIncome",y="CoapplicantIncome",data=df)
```

changing color

```python
sns.relplot(x="ApplicantIncome",y="CoapplicantIncome",data=df,color="pink")
```

with categorical data

```python
sns.relplot(x="LoanAmount",y="ApplicantIncome",hue="Gender",data=df)
```

cat plot=categorical data'

```python
sns.catplot(x="Gender",y="LoanAmount",data=df)
```

```python
sns.catplot(x="Education",y="LoanAmount",data=df)
```

to get everything in a single line use jitter=False

```python
sns.catplot(x="Gender",y="LoanAmount",jitter=False,data=df)
```

with 2 categorical data

```python
sns.catplot(x="Gender",y="LoanAmount",hue="Education",data=df)
```

for getting differnt kind

```python
sns.catplot(x="Education",y="LoanAmount",kind="swarm",data=df)
```

kind=box

```python
sns.catplot(x="Education",y="LoanAmount",kind="box",data=df)
```

kind=violin

```python
sns.catplot(x="Gender",y="LoanAmount",hue="Education",kind="violin",data=df)
```

kind=violin and split=True

```python
sns.catplot(x="Gender",y="LoanAmount",hue="Education",kind="violin",split=True,data=df)
```

kind=bar

```python
sns.catplot(x="Gender",y="LoanAmount",hue="Education",kind="bar",data=df)
```

kind=point

```python
sns.catplot(x="Gender",y="LoanAmount",hue="Education",kind="point",data=df)
```
